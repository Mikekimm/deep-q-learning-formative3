```python
gamma_values = [0.999, 0.99, 0.98, 0.95, 0.90]
batch_values = [8, 16, 32, 64, 128]
```

## Optional: watch reward curves live while training runs

Run this once before starting the loops below to open an inline
TensorBoard dashboard -- it updates live as each run trains, so you can
watch reward trends without waiting for a run to finish. (This shows the

```python
%load_ext tensorboard
%tensorboard --logdir results/tb_logs
```

```python
results = []
for i, gamma in enumerate(gamma_values, start=1):
    row = train_one_run(
        overrides={"gamma": gamma},
        run_name=f"memberB_gamma_{i:02d}",
        member="MemberB",
    )
    results.append(row)
```

```python
for i, batch in enumerate(batch_values, start=1):
    row = train_one_run(
        overrides={"batch_size": batch},
        run_name=f"memberB_batch_{i:02d}",
        member="MemberB",
    )
    results.append(row)
```
